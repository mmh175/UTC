# Photon Splitting at Level 1 Beam Splitter (1A BS)

We consider two photon pairs arriving:

- **Early pair**: `H`, `V` (150 ps apart)
- **Late pair**: `h`, `v` (150 ps apart)
- **Time gap between pairs**: 10 ns → denoted by `-` in listings

Each outcome shows the number of photons going to the **left (→ 2B BS)** and **right (→ 2C BS)** ports.

Let's treat the problem as arranging $n (=4)$ balls in $k (=2)$ boxes. The equation is:

 $$ \text{Number of ways to arrange}= ^{n+k-1}C_{k-1} = \binom{n+k-1}{k-1} $$
Therefore, ways for 4 photons splitting into 2 sides of beam splitters would be: $\binom{4+2-1}{2-1} = 5$
